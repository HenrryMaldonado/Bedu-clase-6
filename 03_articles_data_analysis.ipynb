```python
##Importar librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Lectura de archivo
File_name = 'Datasets/articles_data_homeproject.csv'
df = pd.read_csv(File_name, index_col = 0)
#df = pd.read_csv(File_name)
df.head()
```

```python
#Mostrar encabezado con n elementos
df.head(1)
```

```python
#Visualizar los tipos de datos en un Dataframe
df.dtypes
```

```python
#Visualizar datos nulos en un Dataframe
df.isnull().sum().to_frame()
```

```python
#Análisis de datos
##Saber cuales de los diferentes sitios web(sources) tienen más interacción
Columns = ['source_name',
           'author',
           'engagement_reaction_count',
           'engagement_comment_count',
           'engagement_share_count',
           'engagement_comment_plugin_count'
          ]
df1 = df[Columns].dropna(axis = 0, how = 'any')
df1.head()
```

```python
#Validar que seleccionamos las columnas y que se quitaron los datos nulos
df1.isnull().sum().to_frame()
```

```python
engagement_reaction = df1.groupby('source_name')['engagement_reaction_count'].sum()
engagement_reaction.to_frame()
```

```python
engagement_share = df1.groupby('source_name')['engagement_share_count'].sum()
engagement_share.to_frame()
```

```python
#Extraer nueva serie
highest_source = df1['source_name'].value_counts()
#type(highest_source)
highest_source.to_frame()
```

```python
# Visualización de datos

labels = list(highest_source.index.values)
print(labels)

# Crear gráfica de pastel
plt.pie(highest_source, labels=labels, autopct='%1.1f%%')
plt.title('Fuentes de informacion')
plt.show()
```
